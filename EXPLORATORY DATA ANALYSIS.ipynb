```python
# load the diamonds data into pandas dataframe
import pandas as pd
url='https://raw.githubusercontent.com/tidyverse/ggplot2/main/data-raw/diamonds.csv'
diamonds = pd.read_csv (url)
print(diamonds.shape)
print(diamonds.head()) 
```

```python
# plot the diamonds dataset univariate using ggplot
from plotnine import *
ggplot(data = diamonds) + geom_bar(mapping = aes(x = 'color'))
```

```python
# use histogram to examine the feature "carat" with the width of the bins of 0.2 
ggplot(data = diamonds) + geom_histogram(mapping = aes(x = 'carat'), binwidth = 0.2)
```

```python
# use histogram to examine the feature "carat" with the width of the bins of 0.1 
ggplot(data = diamonds) + geom_histogram(mapping = aes(x = 'carat'), binwidth = 0.1)
```

```python
# pick subsamples of dataset on carat column and create two samples from the "carat" data series
import numpy as np
np.random.seed(22)
sample1 = diamonds['carat'].sample(n=1000,replace=False)
sample2 = diamonds['carat'].sample(n=1000,replace=False)
```

```python
print(sample1)
print(sample2)
```

```python
sample1_df = sample1.to_frame()
sample2_df = sample2.to_frame()
```

```python
# create the histogram fot the first sample
ggplot(data = sample1_df) + geom_histogram(mapping = aes(x = 'sample1', binwidth = 0.1))
```

```python
# create the histogram fot the second sample
ggplot(data = sample2_df) + geom_histogram(mapping = aes(x = 'sample2', binwidth = 0.1))
```

```python
# plot the boxplot to create the visualisation for the "carat" feature
ggplot(data = diamonds) + geom_boxplot(mapping = aes(x='color',y = 'carat'))
```

```python
# plot the boxplot to create the visualisation for the "carat" feature, adding red outlier
ggplot(data = diamonds) + geom_boxplot(mapping = aes(x='color',y = 'carat'),outlier_colour="red")
```

```python
# create a scatter plot mapping the two features
ggplot(data = diamonds) + geom_point(mapping = aes(x = 'carat', y = 'price'))
```

```python
# use different colours to map the cut quality of the diamond onto the scatter plot
ggplot(data = diamonds) + geom_point(mapping = aes(x = 'carat', y = 'price', color = 'cut'))
```

```python
#  identify the spread of observations between category combinations of cut and clarity
ggplot(data = diamonds) + geom_count(mapping = aes(x = 'cut', y= 'clarity'))
# notice that the larger nodes signify the most frequent combinations
```

**PART.2**

```python
# convert categorial data to numeric for corelation computation
# colorNumeric = diamonds['color'].cat.codes
diamonds['color']= diamonds['color'].astype('category')
diamonds['color'] = diamonds['color'].cat.codes
```

```python
diamonds['clarity'] = diamonds['clarity'].astype('category')
diamonds['clarity'] = diamonds['clarity'].cat.codes
diamonds['cut'] = diamonds['cut'].astype('category')
diamonds['cut'] = diamonds['cut'].cat.codes
```

```python
# create a correlation matrix, showing the pearson correlation coefficients between the features,we can use the corr function of the pandas dataframe
corr = diamonds.corr(method = "pearson")
corr
```

```python
# import data visualisation (seaborn) andmatplotlib
import seaborn as sns
import matplotlib.pyplot as plt
heatmap = sns.heatmap(corr, annot=True)
plt.figure(figsize=(9, 4))
plt.show()
```
