**Creating dictionary**

```python
# create a dictionary where the key of each entry is 'Type Mark' (D11, D17, D47, etc.) and the value is a count of the doors in that category.
door_counts = defaultdict(int) # create empty dictionary for loop to iterate over all the rows in df looking at the contents of the 'Type Mark' column
for door in df.loc[:,'Type Mark']:
  door_counts[door] +=1
# display the dictionary 
door_counts.items()
```

**Setting up the data for the plot**

```python
# setting up the data for the plot
type_mark = list(door_counts.keys())
num = list(door_counts.values())     

print ("List of door types: ", type_mark)
print ("List of door counts:",num)
```

**Bar chart**

```python
# draw a bar chart
plt.bar(type_mark, num)
plt.show()
```

**Embellishing the plot**

```python
# add a title and some axis labels to the chart
plt.xlabel("Door Type Mark")
plt.ylabel("No. of doors")
plt.title("Door numbers in the Villa Tugendhat by type")
plt.bar(type_mark, num)
plt.show()
```

**Sorting things and list comprehension**

```python
# create a sorted list from the dictionary
sorted_list = sorted(door_counts.items())
type_mark=[door[0] for door in sorted_list] # create a list of all the first elements
num=[door[1] for door in sorted_list] # create a list of all the second elements
print(type_mark)
print(num)
```

**Plotting the chart**

```python
# plot the chart with the sorted data
plt.xlabel("Door Type Mark")
plt.ylabel("No. of doors")
plt.title("Door numbers in the Villa Tugendhat by type")
plt.bar(type_mark, num)
plt.show()
```

**Sorting dictionaries**

```python
# import the operator library to sort the dictionary 
import operator
sorted_by_type = sorted(door_counts.items(), key=operator.itemgetter(0))
sorted_by_value = sorted(door_counts.items(), key=operator.itemgetter(1))

print(sorted_by_type)
print(sorted_by_value)
```

**Shortcuts**

```python
# use the new sorted dictionaries direclty in the plot function, sorted by type
plt.bar([door[0] for door in sorted_by_type],[door[1] for door in sorted_by_type])
plt.show()
```

```python
# use the new sorted dictionaries direclty in the plot function, sorted by value
plt.bar([door[0] for door in sorted_by_value],[door[1] for door in sorted_by_value])
plt.show()
```

**Counting doors by level**

```python
# count doors by level
floors = defaultdict(int)
for floor in df.loc[:,'Level']:
  floors[floor]+=1
floors.items()
```

**Plotting the results**

```python
# plot the results
plt.bar(list(floors.keys()),list(floors.values()))
plt.show()
```

**Plotting with groupby()**

```python
# number of unique door types per level
df.groupby('Level')['Type Mark'].nunique().plot(kind='bar')
plt.show()
```

```python
# number of door types per level
df.groupby(['Level','Type Mark']).size().unstack().plot(kind='bar')
plt.show()
```

**The plotnine library**

```python
# chart plot using the plotnine library
from plotnine import *
ggplot(data = df) + aes(x = 'Type Mark' ) + geom_bar ()
```

```python
plt.plot(data['Year'],data['Urban population'])
# setting x & y axises labels
plt.xlabel('Year')
plt.ylabel('Urban population')
# setting plot title
plt.title('UK Urban Population')
plt.show()
```

**Plotting multiple data into one chart**

```python
# first plot Urban population
plt.plot(data['Year'],data['Urban population'])
# second plot Rural population on the same figure
plt.plot(data['Year'],data['Rural population'])

# set x axis label and title
plt.xlabel('Year')
plt.title('Urban & Rural population comparison of the UK')
plt.show()
```

```python
# first plot Urban population
plt.plot(data['Year'],data['Urban population'], label="Urban population")
# then plot Rural population on the same figure
plt.plot(data['Year'],data['Rural population'], label="Rural population")

# set x axis label and title
plt.xlabel('Year')
plt.title('Urban & Rural population comparison of the UK')

# adding legend to current plot
plt.legend()
plt.show()
```

**Filtering data**

```python
# create a mask for year >= 2002
mask = data['Year'] >= 2002

# apply mask to the data and assign filtered data to a new variable
masked_data = data[mask]

# plot the previous figure with the masked data
plt.plot(masked_data['Year'],masked_data['Urban population'], label="Urban Population")
plt.plot(masked_data['Year'],masked_data['Rural population'], label="Rural Population")
plt.legend()
plt.xlabel('Year')
plt.title('Urban vs Rural population of the UK after 2002')
plt.show()
```

### **3**

**Loading Data**

```python
# load the data
df = pd.read_csv('jobs.csv')
```

```python
# import the plotly library and initialize for using with the notebook
from plotly.offline import iplot, init_notebook_mode
import plotly.graph_objects as go

init_notebook_mode(connected=True)

# this function needed to make Colab work with plotly, see: https://stackoverflow.com/questions/47230817/plotly-notebook-mode-with-google-colaboratory
def configure_plotly_browser_state():
  import IPython
  display(IPython.core.display.HTML('''
        <script src="/static/components/requirejs/require.js"></script>
        <script>
          requirejs.config({
            paths: {
              base: '/static/base',
              plotly: 'https://cdn.plot.ly/plotly-latest.min.js?noext',
            },
          });
        </script>
        '''))
```

**Visualizing the Data**

```python
# visualise the data
configure_plotly_browser_state() # this line needed to make plotly play nice with Colab
fig = go.Figure(data=go.Scatter(x=df["Date"], y=df["North East"]))
iplot(fig)
```

```python
# add circle markers on the line to improve the appearance of the visualization
configure_plotly_browser_state() # this line needed to make plotly play nice with Colab
fig = go.Figure(data=go.Scatter(x=df["Date"], y=df["North East"], mode="lines+markers"))
iplot(fig)
```

**Annotate the plot**

```python
# annotate the plot
configure_plotly_browser_state() # this line needed to make plotly play nice with Colab
fig = go.Figure(
    data=go.Scatter(
        x=df["Date"],
        y=df["North East"],
        text=df["North East"],
        mode="lines+markers+text",
        textposition="top center",
    )
)
iplot(fig)
```

```python
# annotate the minimum and maximum number of jobs in the North East by creating two annotations
configure_plotly_browser_state() # this line needed to make plotly play nice with Colab
min_annotation = go.layout.Annotation(
    x=df["North East"].idxmin(),  # the index of the minimum value
    y=df["North East"].min(),  # minimum value of `North East` column
    text="Minimum",  # annotation text
    ay=-70,  # margin for annotation in y-axis
    ax=0,  # margin for annotation in x-axis
    showarrow=True,  # whether or not showing and arrow head
    arrowhead=7,  # change of arrow head, ranging from 0-8
)

max_annotation = go.layout.Annotation(
    x=df["North East"].idxmax(),  # the index of the maximum value
    y=df["North East"].max(),  # maximum value of `North East` column
    text="Maximum",  # annotation text
    ay=-40,  # margin for annotation in y-axis
    ax=0,  # margin for annotation in x-axis
    showarrow=True,  # whether or not showing and arrow head
    arrowhead=1,  # change of arrow head, ranging from 0-8
)

fig.update_layout(annotations=[min_annotation , max_annotation])
iplot(fig)
```

```python
configure_plotly_browser_state() #This line needed to make plotly play nice with Colab
min_annotation = go.layout.Annotation(
    x=df["North East"].idxmin(),  # returns the index with of the minimum value
    y=df["North East"].min(),  # returns the minimum value
    text="Minimum Value",
    showarrow=True,
    arrowhead=3,  # arrowhead type, ranging 0-8
    ay=-100,  # margin in y axis for the annotation
)

max_annotation = go.layout.Annotation(
    x=df["North East"].idxmax(),  # returns the index with of the maximum value
    y=df["North East"].max(),  # returns the maximum value
    text="Maximum Value",
    showarrow=True,
    arrowhead=8,  # arrowhead type, ranging 0-8
    ay=-60,  # margin in y axis for the annotation
)
fig = go.Figure(
    data=go.Scatter(
        x=df["Date"],
        y=df["North East"],
        text=df["North East"],
        mode="lines+markers+text",
        textposition="top center",
    ),
    layout=go.Layout(annotations=[min_annotation, max_annotation]),
)

iplot(fig)
```

```python
 # add a title and axis labels to our plot
 configure_plotly_browser_state() # this line needed to make plotly play nice with Colab
fig = go.Figure(
    data=go.Scatter(
        x=df["Date"], y=df["North East"], text=df["North East"], mode="lines+markers"
    ),
    layout=go.Layout(
        title="Number of Professional, scientific and technical activities total workforce jobs",
        xaxis_title="Year",
        yaxis_title="Number of total workforce jobs",
    ),
)
iplot(fig)
```

```python
configure_plotly_browser_state() #This line needed to make plotly play nice with Colab
fig = go.Figure(
    layout=go.Layout(
        title="Number of Professional, scientific and technical activities total workforce jobs",
        xaxis_title="Year",
        yaxis_title="Number of total workforce jobs",
        width=1000,
        height=600
    )
)

for col in df.columns[1:]:  # here col represents the column name
    fig.add_trace(go.Scatter(x=df["Date"], y=df[col], mode="lines+markers", name=col))


iplot(fig)
```
